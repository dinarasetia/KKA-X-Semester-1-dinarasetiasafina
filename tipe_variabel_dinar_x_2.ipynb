#assigment

```python
#single assigment
bb_fazel = 55
bb_dani = 60

print(bb_dani)
print(bb_fazel)
```

```python
#multiple assigment
bb_fazel, bb_dani = 55,60
print(bb_dani)
print(bb_fazel)
```

```python
#variabel dan nilai variabel harus sama banyak
bb_fazel, bb_dani = 55, 60, 70
print(bb_dani)
print(bb_fazel)
```

```python
nama_siswa_1, nama_siswa_2 = "deni","doni"

print(nama_siswa_1)
print(nama_siswa_2)
```

```python
bb_fazel, bb_dani, bb_adli = 55,60
print(bb_dani)
print(bb_fazel)
```

```python
#chain assegment
bb_hasbi = bb_dani = 60
print(bb_hasbi)
print(bb_dani)
```
